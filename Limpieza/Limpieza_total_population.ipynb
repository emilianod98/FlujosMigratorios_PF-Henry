```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
by_sex_destination  = pd.read_excel("Datasets/undesa_pd_2020_ims_stock_by_age_sex_and_destination.xlsx", sheet_name="Table 2", skiprows=10)
```

```python
by_sex_destination
```

```python
columnas_modificar = by_sex_destination.columns[6:23]
nueva_palabra = 'Both'
nuevos_titulos = [nueva_palabra + ' ' + columna for columna in columnas_modificar]
by_sex_destination.rename(columns=dict(zip(columnas_modificar, nuevos_titulos)), inplace=True)


by_sex_destination.head(5)
```

```python
columnas_modificar = by_sex_destination.columns[23:40]
nueva_palabra = 'Male'
nuevos_titulos = [nueva_palabra + ' ' + columna for columna in columnas_modificar]
by_sex_destination.rename(columns=dict(zip(columnas_modificar, nuevos_titulos)), inplace=True)


by_sex_destination.head(5)
```

```python
columnas_modificar = by_sex_destination.columns[40:57]
nueva_palabra = 'Female'
nuevos_titulos = [nueva_palabra + ' ' + columna for columna in columnas_modificar]
by_sex_destination.rename(columns=dict(zip(columnas_modificar, nuevos_titulos)), inplace=True)


by_sex_destination.head(5)
```

```python
columnas_modificar = by_sex_destination.columns[22:]
by_sex_destination.rename(columns=lambda x: x[:-2] if x in columnas_modificar else x, inplace=True)

by_sex_destination.head(5)
```

```python
by_sex_destination.replace("..", 0, inplace=True)
```

```python
by_sex_destination = by_sex_destination.drop(["Index", "Notes", "Location code", "Type of data"], axis=1)

by_sex_destination.head()
```

```python
columna_inicial = 'Both 0-4'

columnas_redondear = by_sex_destination.columns[by_sex_destination.columns.get_loc(columna_inicial):]
by_sex_destination[columnas_redondear] = by_sex_destination[columnas_redondear].round().astype(int)

by_sex_destination.head(5)
```

```python
fila_migraciones = by_sex_destination.iloc[6]

# Obtener los valores de las migraciones
valores_migraciones = fila_migraciones.values[2:18]

# Crear el gráfico de barras
columnas_migraciones = by_sex_destination.columns[2:18]
plt.bar(columnas_migraciones, valores_migraciones)
plt.xlabel('Edad')
plt.ylabel('Migraciones')
plt.title('Migraciones por edad')
plt.xticks(rotation=90)

plt.show()
```

```python
fila_migraciones = by_sex_destination.iloc[6]

valores_migraciones = fila_migraciones.values[19:35]
columnas_migraciones = by_sex_destination.columns[19:35]

plt.bar(columnas_migraciones, valores_migraciones)
plt.xlabel('Edad')
plt.ylabel('Migraciones')
plt.title('Migraciones por edad')
plt.xticks(rotation=90)

plt.show()
```
