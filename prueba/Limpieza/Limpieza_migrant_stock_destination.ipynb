```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
by_sex_destination  = pd.read_excel("Datasets/undesa_pd_2020_ims_stock_by_sex_and_destination.xlsx", sheet_name="Table 1", skiprows=10)
```

```python
by_sex_destination
```

```python
columnas_años = by_sex_destination.columns[5:12]
diccionario_mapeo = {col: f'Both {col}' for col in columnas_años}
by_sex_destination = by_sex_destination.rename(columns=diccionario_mapeo)

by_sex_destination.head(5)
```

```python
columnas_años = by_sex_destination.columns[12:19]
diccionario_mapeo = {col: f'Male {col}' for col in columnas_años}
by_sex_destination = by_sex_destination.rename(columns=diccionario_mapeo)

by_sex_destination.head(5)
```

```python
columnas_años = by_sex_destination.columns[19:30]
diccionario_mapeo = {col: f'Female {col}' for col in columnas_años}
by_sex_destination = by_sex_destination.rename(columns=diccionario_mapeo)

by_sex_destination.head(5)
```

```python
columnas_modificar = by_sex_destination.columns[12:]
by_sex_destination.rename(columns=lambda x: x[:-2] if x in columnas_modificar else x, inplace=True)

by_sex_destination.head(5)
```

```python
by_sex_destination.replace("..", 0, inplace=True)
```

```python
by_sex_destination = by_sex_destination.drop(["Unnamed: 0", "Notes", "Location code", "Type of data"], axis=1)

by_sex_destination.head()
```

```python
fila_migraciones = by_sex_destination.iloc[0]

# Obtener los valores de las migraciones
valores_migraciones = fila_migraciones.values[1:8]

# Crear el gráfico de barras
columnas_migraciones = by_sex_destination.columns[1:8]
plt.bar(columnas_migraciones, valores_migraciones)
plt.xlabel('Year')
plt.ylabel('Migration Stock')
plt.title('Migration Stock per Year(Both)')
plt.xticks(rotation=90)

plt.show()
```

```python
fila_migraciones = by_sex_destination.iloc[0]

valores_migraciones = fila_migraciones.values[8:15]
columnas_migraciones = by_sex_destination.columns[8:15]

plt.bar(columnas_migraciones, valores_migraciones)
plt.xlabel('Year')
plt.ylabel('Migration Stock')
plt.title('Migration Stock per Year(Male)')
plt.xticks(rotation=90)

plt.show()
```

```python
fila_migraciones = by_sex_destination.iloc[0]

valores_migraciones = fila_migraciones.values[15:22]
columnas_migraciones = by_sex_destination.columns[15:22]

plt.bar(columnas_migraciones, valores_migraciones)
plt.xlabel('Year')
plt.ylabel('Migration Stock')
plt.title('Migration Stock per Year(Female)')
plt.xticks(rotation=90)

plt.show()
```
