```python
# Filtrar filas donde 'Income_Group' sea diferente de 'Agregados'
filtrado_df = df.query("Income_Group != 'Agregados'")

# Seleccionar las columnas de interés: 'Income_Group', '2000', '2010', '2020'
columnas_seleccionadas = ['Income_Group', '1980', '2000', '2010', '2020']
resultado = filtrado_df[columnas_seleccionadas]
totales = resultado.groupby('Income_Group').sum()
print(totales)
```

```python
# Filtrar filas donde 'Income_Group' sea diferente de 'Agregados'
filtrado_df = df.query("Region != 'NaN'")

# Seleccionar las columnas de interés: 'Income_Group', '2000', '2010', '2020'
columnas_seleccionadas = ['Region',  '2010', '2020']
resultado = filtrado_df[columnas_seleccionadas]
totales = resultado.groupby('Region').sum()
print(totales)
```
