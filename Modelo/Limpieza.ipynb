```python
import pandas as pd
```

```python
homicidios = pd.read_csv("Datasets/homicides100K.csv")
homicidios.head()
```

```python
homicidios = homicidios.sort_values(by=["Country Name", "Year"])
homicidios = homicidios.reset_index(drop=True)
```

```python
homicidios = homicidios.drop(["Country Code", "Tasa por cada 100K"], axis=1)
homicidios.head()
```

```python
remesas = pd.read_csv("Datasets/remesas.csv")
remesas.head()
```

```python
remesas = remesas.sort_values(by=["Country Name", "Year"])
remesas = remesas.reset_index(drop=True)
```

```python
remesas = remesas.drop("Country Code", axis=1)
remesas.head()
```

```python
remesas.rename(columns={'Value': 'Remesas'}, inplace=True)
```

```python
Dataset = homicidios.merge(remesas, on=["Country Name", "Year"])
Dataset.head()
```

```python
migracion = pd.read_csv("Datasets/Migracion.csv")
migracion.rename(columns={'Country': 'Country Name'}, inplace=True)
```

```python
rango_columnas = migracion.columns[2:38]
migracion = migracion.drop(columns=rango_columnas)
rango_columnas = migracion.columns[3:7]
migracion = migracion.drop(columns=rango_columnas)
migracion.head()
```

```python
Dataset = Dataset.merge(migracion, on=["Country Name", "Year"])
Dataset.head(20)
```

```python
grouped = Dataset.groupby('Country Name')
Dataset['Remesas'] = grouped['Remesas'].apply(lambda x: x.fillna(x.min()))
Dataset.head()
```

```python
Dataset = Dataset.sort_values(['Country Name', 'Year'])
grouped = Dataset.groupby('Country Name')
Dataset['Net migration'] = grouped['Net migration'].apply(lambda group: group.interpolate())
Dataset.head()
```

```python
Dataset = Dataset.fillna(0)
Dataset['Expulsor'] = Dataset['Net migration'].apply(lambda x: 1 if x < 0 else 0)
Dataset['Receptor'] = Dataset['Net migration'].apply(lambda x: 0 if x < 0 else 1)
Dataset.head(30)
```

```python
Dataset.to_csv('Dataset_modelo.csv', index=False)
```
