```python
import pandas as pd
```

```python
Poblacion = pd.read_csv('Datasets/Poblacion_total.csv', skiprows=4)
Poblacion.head()
```

```python
Poblacion = Poblacion.drop(["Indicator Name", "Indicator Code"], axis=1)
rango_columnas = Poblacion.columns[2:63]
Poblacion = Poblacion.drop(columns=rango_columnas)
Poblacion = Poblacion.drop(["Unnamed: 66"], axis=1)
Poblacion.head()
```

```python
Poblacion.columns = ['Country', 'Country Code', 'Total poblation']
Poblacion.head()
```

```python
Migracion = pd.read_csv("Datasets/Migracion.csv")
Migracion.head()
```

```python
rango_columnas = Migracion.columns[2:38]
Migracion = Migracion.drop(columns=rango_columnas)
Migracion.head()
```

```python
rango_columnas = Migracion.columns[3:7]
Migracion = Migracion.drop(columns=rango_columnas)
Migracion.head()
```

```python
Migracion = Migracion[Migracion['Year'] == 2020]
```

```python
Migracion = Migracion.drop("Year", axis=1)
Migracion.head()
```

```python
Dataset = pd.merge(Poblacion, Migracion, on='Country', how='inner')
```

```python
Dataset
```

```python
PBI = pd.read_csv("Datasets/PBI_per_capita.csv", skiprows=4)
PBI.head()
```

```python
rango_columnas = PBI.columns[2:65]
PBI = PBI.drop(columns=rango_columnas)
PBI = PBI.drop("Unnamed: 66", axis=1)
PBI.head()
```

```python
PBI.columns = ['Country', 'Country Code', 'PBI per capita']
```

```python
PBI = PBI.drop("Country", axis=1)
```

```python
Dataset = pd.merge(Dataset, PBI, on='Country Code', how='inner')
Dataset.head()
```

```python
Remesas = pd.read_csv("Datasets/Remesas.csv", skiprows=4)
Remesas.head()
```

```python
rango_columnas = Remesas.columns[2:65]
Remesas = Remesas.drop(columns=rango_columnas)
Remesas = Remesas.drop("Unnamed: 66", axis=1)
Remesas.head()
```

```python
Remesas.columns = ['Country', 'Country Code', 'Remesas']
Remesas.head()
```

```python
Remesas = Remesas.drop("Country", axis=1)
```

```python
Dataset = pd.merge(Dataset, Remesas, on='Country Code', how='inner')
Dataset.head()
```

```python
Homicidios = pd.read_csv("Datasets/Homicidios.csv", skiprows=4)
Homicidios.head()
```

```python
rango_columnas = Homicidios.columns[2:64]
Homicidios = Homicidios.drop(columns=rango_columnas)
Homicidios = Homicidios.drop("Unnamed: 66", axis=1)
Homicidios = Homicidios.drop("2021", axis=1)
Homicidios.head()
```

```python
Homicidios.columns = ['Country', 'Country Code', 'Homicidios cada 100.000 Hab']
Homicidios.head()
```

```python
Homicidios = Homicidios.drop("Country", axis=1)
```

```python
Dataset = pd.merge(Dataset, Homicidios, on='Country Code', how='inner')
Dataset.head()
```

```python
new_order = ['Country', 'Country Code', 'Net migration', 'Total poblation', 'PBI per capita', 'Remesas', 'Homicidios cada 100.000 Hab']
Dataset = Dataset[new_order]
Dataset.head()
```

```python
Dataset
```

```python
Dataset = Dataset.fillna(0)
```

```python
Dataset["Homicidios"] = Dataset["Homicidios cada 100.000 Hab"]
Dataset = Dataset.drop("Homicidios cada 100.000 Hab", axis=1)
Dataset.head()
```

```python
numeric_columns = Dataset.select_dtypes(include=['float64', 'int64'])
Dataset[numeric_columns.columns] = numeric_columns.round(2)
Dataset = Dataset.loc[Dataset["Net migration"] != 0]
```

```python
Dataset.to_csv('Dataset_modelo.csv', index=False)
```
